```python
def count(list):
    total = 0
    for x in list:
        if x < 20:
            total = total + 1
    return total
            
```

```python
numbers = [1,2,3,4,55,22]
count(numbers)
```

```python
count(numbers)
```

```python
prices = {
    "spaghetti" : 14,
    "pasta" : 4 , 
    "pizza" : 3  
}

quantity = {
    "spaghetti" : 2,
    "pasta" : 4,
    "pizza" :1 
}
```

```python
spent = 0
for i in prices:
    spent = spent + prices[i] * quantity[i]
print(spent)
```
